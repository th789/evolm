```python
import pandas as pd
```

```python
#load json file
domain_data = pd.read_json("evolm/common/scripts/ffw_domain_info_mysubset_20BT.json", lines=True)

print(domain_data.shape)
print(domain_data.columns)
```

```python
#for each domain, examine #tokens desired vs #tokens obtained

domains = domain_data['domain'].unique()

for i, domain in enumerate(domains):
    print(i, domain)

    n_tokens_desired = domain_data[domain_data["domain"] == domain]["n_tokens_desired"].item()
    print(f'   # tokens desired: {n_tokens_desired:,}')

    n_tokens_obtained = domain_data[domain_data["domain"] == domain]["n_tokens"].item()
    print(f'   # tokens obtained: {n_tokens_obtained:,}')

    print(f'   Difference: {n_tokens_obtained-n_tokens_desired:,}')
    print()
```

```python
#check domain of interest
domain_of_interest = 'mathematics'
domain_data[domain_data['domain'] == domain_of_interest]
```

```python
#count total number of tokens in my subset
total_n_tokens_desired = domain_data['n_tokens_desired'].sum()
print(f'Total #tokens desired: {total_n_tokens_desired:,}')

total_n_tokens_obtained = domain_data['n_tokens'].sum()
print(f'Total #tokens obtained: {total_n_tokens_obtained:,}')

print(f'Difference: {total_n_tokens_obtained - total_n_tokens_desired:,}')
```

```python
#other stats
print('Total # json files: ', domain_data['n_json_files'].sum())

total_n_docs = domain_data['n_docs'].sum()
print(f'Total # docs: {total_n_docs:,}')
```
